```python
import os
import numpy as np # type: ignore
import librosa.display # type: ignore
import matplotlib.pyplot as plt # type: ignore
```

```python
# Calculate Spectrogram by using SFTF method
def spectrogram_cal(y,fs):
    
    ms = librosa.feature.melspectrogram(y=y, sr=fs, n_fft=2048, hop_length=128, n_mels=256)
    spectrogram_db = librosa.power_to_db(ms, ref=np.max)
    
    # n_fft = 1024       # Length of FFT window
    # hop_length = 64   # Number of samples between frames
    # win_length = 1024  # Length of the window
    # window = 'hann'    # Windowing function

    # # Compute the STFT
    # spectrogram = librosa.stft(y, n_fft=n_fft, hop_length=hop_length, win_length=win_length, window=window)
    # spectrogram_db = librosa.amplitude_to_db(np.abs(spectrogram))  # Convert to dB
    
    return spectrogram_db
```

```python
# Load the Drone audio file
audio_file = './dataset/Sound/Drone/Drone_16.wav'  # Replace with your audio file path
y, fs = librosa.load(audio_file)  # , duration=20)
timesDuration = librosa.get_duration(y=y, sr=fs)

print(f'Sampling Rate: {fs} Hz')
print(f'Audio Duration: {timesDuration:.0f} seconds')

plt.figure(figsize=(10, 8))

# ------- Plot Spectrogram ---------
spectrogram_db = spectrogram_cal(y,fs)

plt.subplot(2, 1, 2)
librosa.display.specshow(spectrogram_db, sr=fs, x_axis='time', y_axis='linear',cmap='viridis')
#cmap = 'viridis', 'plasma', 'inferno', 'magma', 'cividis'
plt.colorbar(format='%+2.0f dB')
plt.title(f'Shpae of spectrogram :  {spectrogram_db.shape}')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.grid()

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

```

```python
sound_dataset_dir = "./dataset/Sound/No_Drone"
path_save_spectrogram = "./dataset/train/No_Drone"

count = 0   # for count number to save img
for audio_file in os.listdir(sound_dataset_dir):
        audio_path = os.path.join(sound_dataset_dir, audio_file)
        y, sr = librosa.load(audio_path)  # Load audio and limit to 3 seconds    
        sp_db = spectrogram_cal (y,sr)
        
        plt.figure()
        plt.subplot(1,1,1)
        plt.subplots_adjust(left=0,right=1,bottom=0,top=1)
        
        librosa.display.specshow(sp_db,sr=sr,cmap='viridis')
        
        pathSaveImg = path_save_spectrogram + "/" + str(count)
        plt.savefig(pathSaveImg)
        plt.close()
        
        count = count + 1
        print(f'{count} : {pathSaveImg}')
```
